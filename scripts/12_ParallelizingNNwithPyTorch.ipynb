Intro to PyTorch

```python
import torch
import numpy as np

np.set_printoptions(precision=3)
```

```python
a = [1,2,3]
b = np.array([4,5,6], dtype=np.int32)

print(f"a: {a}, {type(a)}\nb: {b} {type(b)}")

t_a = torch.tensor(a)
t_b = torch.tensor(b)
t_b2 = torch.from_numpy(b)

print(f"t_a: {t_a}, {type(t_a)}\nt_b: {t_b} {type(t_b)}\nt_b2: {t_b2} {type(t_b2)}")
print(f"t_a: {t_a}, shape: {t_a.shape}")
```

```python
t_ones = torch.ones(2,3)
print(f"tensor:\n{t_ones},\nshape: {t_ones.shape}")

rand_tensor = torch.rand(2,3)
print(rand_tensor)
```

```python
t_a_new = t_a.to(torch.int32)
print(f"tensor:\n{t_a},\ndtype: {t_a.dtype}")
print(f"tensor:\n{t_a_new},\ndtype: {t_a_new.dtype}")
```

```python
# Transposing a tensor
t = torch.rand(3,5)
t_tr = t.transpose(0, 1) # You have to specify which dimensions have to be transpose. Tensors can have more than 2 dimensions.

print(f"t: {t}\nt_tr: {t_tr}\n{t.shape} ---> {t_tr.shape}")
```

```python
# Reshaping a tensor
t = torch.zeros(30)
t_reshape = t.reshape(5,-1)

print(f"t:\n{t}\nt_reshape:\n{t_reshape}")

print(f"{t.shape} ---> {t_reshape.shape}")
```

```python
t = torch.zeros(1,2,1,4,1)
t_sqz = torch.squeeze(t,2)
print(f"t: {t},\nt_sqz: {t_sqz}")
print(t.shape, ' --> ', t_sqz.shape)
```

```python
torch.manual_seed(1)
t1 = 2 * torch.rand(5,2) - 1
t2 = torch.normal(mean=0, std=1, size=(5,2))

print(f"t1:\n{t1},\nt2:\n{t2}")
```

```python
t3 = torch.multiply(t1, t2) # eementwise multiplication
print(t3)
```

```python
print(f"Mean function: {torch.mean(t1, axis=0)}\nMean method: {t1.mean(axis=0)}")
print("*"*20)
print(f"Sum function: {torch.sum(t1, axis=0)}\nSum method: {t1.sum(axis=0)}")
print("*"*20)
print(f"Std function: {torch.std(t1, axis=0)}\nStd method: {t1.std(axis=0)}")
print("*"*50)
print(f"Mean function: {torch.mean(t1, axis=1)}\nMean method: {t1.mean(axis=1)}")
print("*"*20)
print(f"Sum function: {torch.sum(t1, axis=1)}\nSum method: {t1.sum(axis=1)}")
print("*"*20)
print(f"Std function: {torch.std(t1, axis=1)}\nStd method: {t1.std(axis=1)}")
```

```python
print(f"Matrix multiplication (result 5x5):\n{t1.matmul(t2.transpose(0,1))}\n")
print(f"Matrix multiplication (result 2x2):\n{t1.transpose(0,1).matmul(t2)}")
```

```python
print(f"L^p norm (for p=2 and 2nd dim):\n{torch.linalg.norm(t1, ord=2, dim=1)}")
print(f"L^p norm (for p=2 and 1st dim):\n{torch.linalg.norm(t1, ord=2, dim=0)}")

print("To check if it works properly:\n")
print(f"{np.sqrt(np.sum(np.square(t1.numpy()), axis=1))}")
# print(np.square(t1.numpy()))
```

```python
torch.manual_seed(1)
t_x = torch.rand([4,3], dtype=torch.float32)
t_y = torch.arange(4)

print(f"t_x:\n{t_x},\nt_y:\n{t_y}")
```

```python
from torch.utils.data import Dataset
class JoinDataset(Dataset):
    def __init__(self, x, y) -> None:
        self.x = x
        self.y = y
    
    def __len__(self):
        return len(self.x)
    
    def __getitem__(self, idx):
        return self.x[idx], self.y[idx]

joint_dataset = JoinDataset(t_x, t_y)
for example in joint_dataset:
    print(f" x: {example[0]}, y: {example[1]}")

print("-"*50)
from torch.utils.data import TensorDataset # but it can be used if the dataset with labels is also a tensor.
joint_dataset2 = TensorDataset(t_x, t_y)
for example in joint_dataset2:
    print(f" x: {example[0]}, y: {example[1]}")
```

```python
# Shuffle, batch, and repeat
torch.manual_seed(1)

data_loader = DataLoader(dataset=joint_dataset2, batch_size=2, shuffle=True)

for i, batch in enumerate(data_loader, 1):
    print(f"batch {i}:\nx: {batch[0]}\ny: {batch[1]}\n")
```

```python
for epoch in range(2):
    print(f"epoch {epoch+1}:")
    for i, batch in enumerate(data_loader, 1):
        print(f"    batch {i}:\n    x: {batch[0]}\n    y: {batch[1]}\n")
```

```python
# Creating a dataset from files on your local storage disk
import pathlib
imgdir_path = pathlib.Path('../images/')
file_list = sorted([str(path) for path in imgdir_path.glob("*.jpg")])
print(file_list)
```

```python
import matplotlib.pyplot as plt
import os
from PIL import Image

fig = plt.figure(figsize=(10,5))
for i, file in enumerate(file_list):
    img = Image.open(file)
    print(f"Image shape: {np.array(img).shape}")
    ax = fig.add_subplot(2,3,i+1)
    ax.set_xticks([]); ax.set_yticks([])
    ax.imshow(img)
    ax.set_title(os.path.basename(file), size=15)
plt.tight_layout()
```

```python
labels = [1 if "dog" in os.path.basename(file) else 0 for file in file_list]
print(labels)
```

```python
import torchvision.transforms as transforms

class ImageDataset(Dataset):
    def __init__(self, file_list, labels, transform=None) -> None:
        self.file_list = file_list
        self.labels = labels
        self.transform = transform

    # img_height, img_width = 80, 120
    # transform = transforms.Compose([
    #     transforms.ToTensor(),
    #     transforms.Resize((img_height, img_width))
    # ])

    def __getitem__(self, index):
        img = Image.open(self.file_list[index])
        if self.transform is not None:
            img = self.transform(img)
        label = self.labels[index]
        return img, label

    def __len__(self):
        return len(self.labels)

img_height, img_width = 80, 120
transform = transforms.Compose([
    transforms.ToTensor(),
    transforms.Resize((img_height, img_width))
])

# img = Image.open(file_list[0])
# transform(img).shape

image_dataset = ImageDataset(file_list, labels, transform)
for file, label in image_dataset:
    print(file, label)
```

```python
fig = plt.figure(figsize=(10,6))

for i, example in enumerate(image_dataset):
    ax = fig.add_subplot(2,3,i+1)
    ax.set_xticks([]), ax.set_yticks([])
    ax.imshow(example[0].numpy().transpose((1,2,0))) # example[0] - only features, base shape: (3, 80, 120), transpose 1st and 2nd dimension to shape: (80, 120, 3)
    # ax.imshow(example[0].numpy().transpose((2,1,0))) # it is rotated imgages by -90
    ax.set_title(f'{example[1]}', size=15)
```
